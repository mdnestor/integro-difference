```python
reload(PyIDE)

growthFunction = PyIDE.RickerGrowthFunction(r=1)

disperalKernel = PyIDE.LaplaceDisperalKernel()

model = PyIDE.IDEModel(growthFunction,
                  disperalKernel)
```

```python
u0 = lambda x : np.heaviside(-x, 1)

sim = PyIDE.IDESimulation(model)

sim.setDomain(0, 200, .01)
sim.setInitialCondition(u0)
sim.setBoundaryCondition("static")


sim.run(150)
sim.plot()
```

```python
sim.boundaryCondition
```

```python
# need a function to estimate wave speed

h = .5

X = sim.domain
ts = sim.timeSeries

W = []

for U in ts:
    
    levelSet = [X[i] for i in range(len(X)) if U[i] > h]
    if len(levelSet)>0:
        x0 = max(levelSet)
    
        W.append(x0)



plt.plot(np.gradient(W))

plt.plot(np.gradient(np.gradient(W)))
```

Try to get the formal estimate now

$$ c^* = \inf_{\lambda>0}\frac{1}{\lambda}\log\left(g'(0)\int k(y)e^{\lambda y}dy\right) $$

$$ g'(u) = \frac{d}{du}ue^{r(1-u)} = -r(u+1)e^{r(1-u)} $$
$$ g'(0) = -re^r $$

$$ \int \frac{1}{2}e^{-|y|/2}e^{\lambda y}dy $$

$$ = \frac{1}{2} \int_{-\infty}^0 e^{y/2}e^{\lambda y}dy + \frac{1}{2} \int_0^\infty e^{-y/2}e^{\lambda y}dy $$

$$ = \frac{1}{2} \int_0^\infty e^{y(-1/2-\lambda)}dy + \frac{1}{2} \int_0^\infty e^{y(-1/2+\lambda)}dy $$

$$ = \frac{1}{2} \frac{1}{-1/2-\lambda} e^{y(-1/2-\lambda)} \mid_0^\infty + \frac{1}{-1/2+\lambda} e^{y(-1/2+\lambda)} \mid_0^\infty $$

$$ = \frac{1}{2} \frac{1}{-1/2-\lambda} (-1) + \frac{1}{-1/2+\lambda} (-1) $$ if $\lambda<1/2$

$$ = \frac{1}{1+2\lambda}+ \frac{1}{1-2\lambda} $$

$$ = \frac{1+2\lambda + 1 - 2\lambda}{1-4\lambda^2} $$

$$ = \frac{2}{1-4\lambda^2} $$

So $$ c^* = \inf_{\lambda>0} \frac{1}{\lambda} \log \left( -re^r\frac{2}{1-4\lambda^2} \right) $$

For $r=1$,

$$ c^* = \inf_{\lambda>0} \frac{1}{\lambda} \log \left(\frac{2e}{4\lambda^2-1} \right) $$
